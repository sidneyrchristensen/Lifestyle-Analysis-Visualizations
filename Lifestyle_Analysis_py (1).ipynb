```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')
```

```python
scatterplt = sns.scatterplot(data = df, x = 'Age', y = 'Sleep Duration', hue = 'Gender', style = 'BMI Category', palette = 'seismic')
scatterplt.set(title = 'Age vs. Sleep Duration')
#scatterplot.title()
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
boxplt = sns.boxplot(x = 'Quality of Sleep', y = 'Physical Activity Level', data = df, palette = 'seismic')
boxplt.set(title = 'Quality of Sleep vs. Physical Activity Level')
```

```python
bmi_boxplt = sns.boxplot(x = 'BMI Category', y = 'Sleep Duration', data = df, palette = "PuRd")
bmi_boxplt.set(title = 'BMI vs. Sleep Duration')
```

```python
lineplt = sns.lineplot(data = df, x = 'Age', y = 'Sleep Duration', hue = "Stress Level")
lineplt.set(title = "Age vs. Sleep Duration")
```

```python
bmi_boxplt = sns.boxplot(x = 'BMI Category', y = 'Sleep Duration', data = df, palette = "PuRd")
bmi_boxplt.set(title = 'BMI vs. Sleep Duration')
```

```python
lineplt = sns.lineplot(data = df, x = 'Age', y = 'Sleep Duration', hue = "Stress Level")
lineplt.set(title = "Age vs. Sleep Duration")
```

```python
df['sleep_disorder_grouped'] = df['Sleep Disorder'].replace({'Sleep Apnea' : 'Sleep Disorder', 'Insomnia' : 'Sleep Disorder'})
plt.figure(figsize=(12, 8))
df.groupby(['Occupation', 'sleep_disorder_grouped']).size().unstack().plot(kind='bar', stacked=False, color = ['#17becf', '#9467bd'])
plt.title("Occupation and Sleep Disorder")
plt.xlabel("")
plt.ylabel("Count")
plt.xticks(rotation=45, ha='right')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
displt = sns.displot(data = df, x='Age', hue='Sleep Disorder', col = 'Gender', kde=True, palette = 'Set1')
plt.xlim(20, 70)
plt.xticks(range(20, 71, 5))
plt.grid(False)
plt.show()
```

```python
df['Gender'] = pd.factorize(df['Gender'])[0]
df['Occupation'] = pd.factorize(df['Occupation'])[0]
df['BMI Category'] = pd.factorize(df['BMI Category'])[0]
df['Blood Pressure'] = pd.factorize(df['Blood Pressure'])[0]
df['Sleep Disorder'] = pd.factorize(df['Sleep Disorder'])[0]

df['Gender'] = df['Gender'].astype(float)
df['Occupation'] = df['Occupation'].astype(float)
df['BMI Category'] = df['BMI Category'].astype(float)
df['Blood Pressure'] = df['Blood Pressure'].astype(float)
df['Sleep Disorder'] = df['Sleep Disorder'].astype(float)

#df = df.drop('sleep_disorder_grouped', axis=1)

correlation_matrix = df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, cmap='spring', annot=False, fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix")
plt.show()
```
